### **1. Descargar librerías y Base de datos**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
poblacion = pd.read_csv("Archivos de estudio\poblacion.csv", encoding='latin1')
vida = pd.read_csv("Archivos de estudio\esperanza_de_vida.csv", encoding='latin1')
supkm2 = pd.read_csv ("Archivos de estudio\hogares_viviendas_superficie.csv", encoding = 'latin1')
```

### **4.1. Informe - Densidad Pobacional Argentina**

Este informe se basa en los datos proporcionados en la base de datos creada **"andemsoc"** y el estudio y análisisde la **"densidad de población expresada en habitantes por km²"**.

En primer lugar, con la máxima expresión, tenemos a Capital Federal con 15345.20 habitantes por km², seguido de la provincia de Tucumán con 81.18  habitantes por km² y en tercer lugar, la provincia de Buenos Aires con una densidad de población de 60.71 habitantes por km².
Esto se puede deber, a que las grandes poblaciones se encomunan en las grandes ciudades universitarias, lo que puede explicar que la provincia de Tucumán se encuentre entre los primeros lugares. A su vez, también se puede deber a que las poblaciones de provincias rurales, turísticas o de bajos recursos, los jóvenes emigren a las grandes ciudades en busca de una mejor escala salarial por la variedad de oportunidades laborales que éstas ofrecen.

En esta misma escala de "densidad", las provincias con menor densidad de población son: Tierra del Fuego con 0.20 habitantes por km² y Santa Cruz con 1.78 habitantes por km². En el caso de Santa Cruz, a pesar de su superficie, muchas de estas zonas no son habitables por su geografía, las pequeñas *"grandes ciudades"* en estas provincias, concentran muchos turistas, por lo que su ocupación habitacional autóctona es pequeña a comparación de las comunidades rurales, aisladas de este circuito.

En términos generales, se observa una gran variabilidad en la densidad de población entre las provincias argentinas, lo que refleja las diferencias en la distribución demográfica a lo largo del país, a pesar de la superficie territorial.

**Informe**

El análisis de la esperanza de vida nos arroja distintos resultados según la variable en estudio. Uno de ellos es la diferencia de esperanza de vida entre mujeres y hombres en argentina, en promedio, podemos encontrarnos con que la diferencia es de 6.29, según la provincia varía la diferencia entre 4 a 6 años de diferencia.
Puede deberse a que los hombres, por una cuestión biológica, sean más propensos a problemas de salud que las mujeres y tienen tendencia a no buscar atención médica inmediata. La mayor parte de Argentina por una cuestion socio cultural, los hombres obtienen una carga más pesadas de trabajo que las mujeres: cargar bultos, ser camioneros, trabajos pesados en la construcción, explotación minera, rurales, entre otros. Lo que provoca un deterioro de su bienestar mas rápido que en el de la mujer.

Por otro lado, realizamos un estudio más general, tomando los valores de esperanza de vida en cuanto al territorio argentino, esto fue para poder observar y localizar qué provincias tienen la mayor y menor esperanza de vida. Así pudimos concretar que las provincias de Chubut, Santa Fe y Santa Cruz tienen los valores más altos de esperanza de vida, mientras que San Juan, Mendoza y Tierra del Fuego tienen los valores más bajos. Ésto puedo ser influenciado por una combinación de factores.

En primer lugar, vale la observación que se esperaba encontrar entre los primeros puestos, provincias con temperaturas templadas a cálidas, con una mayor vegetación y fauna, que serían las que se encuentran al centro y norte argentino, ya que al ser ricos en cuanto comida sana, natural y una vida menos estresante a comparación de las grandes capitales donde hay más consumo de alimentos procesados y rutinas poco saludables como ser Buenos Aires y Capital Federal. Sorprende encontrarnos con que estas zonas, se encuentran en la media de esperanza de vida. Por lo que tuvimos que estudiar otros factores que podrían estar influenciando.

Así, nos encontramos con los servicios de salud y factores ambientales, los cuales, combinados expican por qué una región fría como Chubut y Santa cruz se encuentran entre los primeros lugares, estas son provincias son reconocidas por tener mejor calidad de agua potable y un aire más limpio que al norte argentino, po ejemplo: Salta, Jujuy y Tucumán, que es de público conocimiento que existen regiones, localidades completas, que no cuentan con estos servicios ni una campaña fuerte de concientización del cuidado de salud y cuidado del medio ambiente, lo que impacta directamente en la esperanza de vida.
Santa Fé en cambio, se puede tomar en cuenta que tiene uno de las mejores accesos a los Servicios de Salud (al igual de Córdoba) del país, hasta se puede mencionar que tienen reconocimientos a nivel mundial por ello. Por otro lado las campañas en el cuidado del bienestar personal y concientización del cuidado de los espacios públicos, hizo de esta provincia, ser reconocida por una cultura mucho más saludable en cuanto alimentación, deportes y un gran cuidado con los espacios verdes, hitóricos y culturales. 
Siendo así, podría compararse Santa Fe con Mendoza ya que cuentan con estilos de vida y campañas de bienestar y cuidado muy parecidos, aún así, es una región propensa a cambios climáticos muy fuertes, y no cuentan con un acceso a los servicios de salud pública como Santa Fe, por lo que empobrece el servicio, las personas prefieren curarse solos ya que es un gran esfuerzo económico acceder a una mejor calidad del servicio. Por ello, es posible entender por qué a pesar de ser parecidos se encuentran en extremos opuestos.
En el caso de San Juan, no sorprende que se encuentre en este puesto, ya que, aunque tiene características geográficas diferentes a Chubut o Santa Cruz, con un clima más cálido, y una geografía menos accidentada, la calidad del agua, del aire y del entorno en general no contribuyen a una mejor salud y esperanza de vida.
Por último, Tierra del Fuego, a pesar de su cercanía con los puestos más altos, tiene uno de los valores más bajos. Aquí influyen muchos factores en medidas diferentes. Debido a su clima extremo y una geografía muy pobre en materia prima, hace que el estilo de vida sea aun más laboriosa que en el resto del país. Por lo que en su historia, la migración de un gran porcentaje de la población jóven en busca de un mejor estilo de vida en tierras más fértiles, hace de esta provincia tener una mayor población de gente mayor que jóvenes que puedan dedicarse a la producción y cuidado del adulto mayor. Por lo que este último grupo poblacional, tiene una tendencia a morir más joven que en el resto de Argentina.

##### Esperanza de Vida. Diferencia y Promedio

```python
vida_f = vida.rename (columns = {'varones':'hombres'}, inplace=False)
vida_df=vida_f
vida_df ['diferencia_ev'] = vida_f ['mujeres'] - vida_f ['hombres']
vida_df.sort_values(by=['diferencia_ev'], ascending=False).round(2)
vida_df
```

* **Gráfico Boxplot**

```python
print ("Q1 = "+str(round(np.percentile(vida_df["diferencia_ev"],25),2)))
print ("Q2 = "+str(round(np.percentile(vida_df["diferencia_ev"],50),2)))
print ("Q3 = "+str(round(np.percentile(vida_df["diferencia_ev"],75),2)))
```

Con esta gráfica, podemos corroborar la distriución quinquenal de la esperanza de vida en Agentina.

##### AnDemSoc. Densidad Poblacional

```python
pob_f = poblacion.rename(columns={'poblacion_total':'pob_total', 'poblacion_varones':'pob_hombres', 'poblacion_mujeres':'pob_mujeres'}, inplace=False)
pob_df = pob_f[poblacion['provincia'] != 'Total País']
valores_anio=[2010,2015,2020,2025,2030,2035,2040]
pob_df=pob_df[pob_df['anio'].isin(valores_anio)]
pob_df.sort_values(by=['provincia', 'anio'])
```

```python
supkm2_df=supkm2
supkm2_df = supkm2_df.drop(['provincia_id', 'hogares', 'viviendas_particulares', 'viviendas_particulares_habitadas'],axis=1)
supkm2_df.rename(columns={'superficie_km2':'sup_km2'},inplace=True)
supkm2_df.sort_values(by=['provincia'])
```

```python
andemsoc =pd.merge (vida_df, pob_df)
andemsoc =pd.merge (andemsoc, supkm2_df)
andemsoc ['densidad_pob'] = (andemsoc ['pob_total'] / andemsoc ['sup_km2']).round(2)
andemsoc.sort_values(by=['densidad_pob', 'anio'], ascending=False)
andemsoc.groupby('anio')['densidad_pob']
andemsoc
```

En primer lugar, por medio de un Histograma, buscamos corrovorar que los datos perteneciente al informe de *Esperanza de vida*  sean concluyentes con los datos de la base compuesta para el *Análisis Demográfico y Social*  (AnDemSoc)

```python
plt.figure(figsize=[7,4])
plt.hist(andemsoc['diferencia_ev'])
plt.title('Diferencia Esperaza de Vida. Promedio= ' + str(round(andemsoc['diferencia_ev'].mean(),1)))
plt.axvline (x=andemsoc['diferencia_ev'].mean(), color='black' , linestyle='--')
```

* **Gráfico Histograma**

```python
plt.figure(figsize=[7,4])
plt.hist(andemsoc['pob_total'])
plt.axvline (x=andemsoc['pob_total'].mean(), color='black' , linestyle='--')
```

* **Gráficos de densidad o de densidad de kernel**

```python
plt.figure(figsize=[7,7])
sns.kdeplot(data=andemsoc, x = "pob_total", hue = "provincia")
plt.ylabel("densidad_pob")
```

Tanto el gráfico Boxplot y Kernel, nos demuestran que la densidad en poblacioal parece matener un estandar excluyendo a Capital Federal y Tucuman que marcan valores totalmente atípicos tanto por su superficie y su total de población.
Para comprobar estos datos, repetiremos la ultima gráfica excluyendo estos dos valores atipicos altos y el atípico más bajo, Tierra del Fuego.

```python
andemsoc_f =andemsoc[andemsoc['provincia'] != 'Capital Federal']
andemsoc_f =andemsoc_f[andemsoc['provincia'] != 'Tucumán']
andemsoc_f =andemsoc_f[andemsoc['provincia'] != 'Tierra del Fuego']
plt.figure(figsize=[7,7])
sns.kdeplot(data=andemsoc_f, x = "densidad_pob", hue = "provincia")
plt.ylabel("pob_total")
```

Como se esperaba, la densidad del resto de las provincias se vuelve más legible para su lectura.
